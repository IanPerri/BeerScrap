```python
from bs4 import BeautifulSoup
import requests  
import pandas as pd
import re
```

```python
#Conexión con Mercado Libre
URL = 'https://listado.mercadolibre.com.mx/cerveza-12#D[A:cerveza%2012]'
page = requests.get(URL)
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
find_price = soup.find_all('div',class_="ui-search-item__group__element ui-search-price__part-without-link")
find_price[0]
```

```python
precios=[i.text for i in find_price]
print(len(precios))
precios
```

```python
precios[-2].split('$')[0]
```

```python
find_cervezas = soup.find_all('h2',class_="ui-search-item__title")
find_cervezas[0]
```

```python
cervezas = [i.text for i in find_cervezas]
print(len(cervezas))
cervezas
```

```python
def limpia_precios(lista_precios):
    #precios_final = [lista_precios[i].split('$')[-2] for i in range(len(lista_precios))]
    precios_final = list()
    for i in range(len(lista_precios)):
        text = lista_precios[i].split('$')[0]
        end = text.index(' pesos')
        substring = text[:end]
        precios_final.append(substring)
        
    return(precios_final)
```

```python
precios_cerveza = limpia_precios(precios)
precios_cerveza
```

```python
find_img = soup.find_all('img',class_="ui-search-result-image__element")
find_img[0]
```

```python
img_cerveza = [i.get('data-src') for i in find_img]
print(len(img_cerveza))
img_cerveza
```

```python
find_links = soup.find_all('a', class_ = 'ui-search-item__group__element ui-search-link')
find_links[0]
```

```python
link_cerveza =[ i.get('href') for i in find_links ]
print(len(link_cerveza))
link_cerveza[0]
```

```python
tienda_ml = ['Mercado Libre'] * len(link_cerveza)
len(tienda_ml)
```
